```python
import cv2
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
%matplotlib inline

import time
```

```python
# Görüntüyü yükle
image = cv2.imread('veri/1.jpeg', 0)  # 0, görüntüyü gri seviyede yükler
```

```python
# Histogramı hesapla
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

# Histogramı görselleştir
plt.plot(hist)
plt.xlabel('Piksel Değeri')
plt.ylabel('Frekans')
plt.title('Gri Seviye Histogramı')
plt.show()
```

```python
# Görüntüyü BGR olarak yükle
image_bgr = cv2.imread('veri/1.jpeg')
```

```python
# OpenCV, görüntüleri varsayılan olarak BGR renk düzeninde okur
# Daha sonra renk düzenini RGB'ye dönüştürmek için cv2.cvtColor fonksiyonunu kullanacağız
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

```python
# Renk kanallarını ayrı ayrı hesapla
hist_b = cv2.calcHist([image_rgb], [2], None, [256], [0, 256])
hist_g = cv2.calcHist([image_rgb], [1], None, [256], [0, 256])
hist_r = cv2.calcHist([image_rgb], [0], None, [256], [0, 256])

# Histogramları görselleştir
plt.plot(hist_b, color='blue')
plt.plot(hist_g, color='green')
plt.plot(hist_r, color='red')
plt.xlabel('Piksel Değeri')
plt.ylabel('Piksel Sayısı(Frekans)')
plt.title('Renkli Görüntünün Histogramı')
plt.show()
```

```python
from image_enhancement import image_enhancement
import cv2 as cv
```

```python
image = cv.imread('veri/1.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
ie = image_enhancement.IE(image)
GHE = ie.GHE()
BBHE = ie.BBHE()
QBHE = ie.QBHE(100)
```

```python
# Orijinal ve BPBHE uygulanmış görüntüyü yan yana göster
fig, axes = plt.subplots(2, 4, figsize=(30, 10))
axes[0,0].imshow(image)
axes[0,0].set_title('Orijinal Görüntü')
axes[0,0].axis('off')
axes[0,1].imshow(GHE)
axes[0,1].set_title('GHE ')
axes[0,1].axis('off')
axes[0,2].imshow(BBHE)
axes[0,2].set_title('BBHE')
axes[0,2].axis('off')
axes[0,3].imshow(QBHE)
axes[0,3].set_title('QBHE ')
axes[0,3].axis('off')
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
axes[1,0].plot(hist)
axes[1,0].set_title('Orijinal Görüntü')
hist = cv2.calcHist([GHE], [0], None, [256], [0, 256])
axes[1,1].plot(hist)
axes[1,2].set_title('GHE ')
hist = cv2.calcHist([BBHE], [0], None, [256], [0, 256])
axes[1,2].plot(hist)
axes[1,2].set_title('BBHE')
hist = cv2.calcHist([QBHE], [0], None, [256], [0, 256])
axes[1,3].plot(hist)
axes[1,3].set_title('QBHE ')
plt.show()
```
